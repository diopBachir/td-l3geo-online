# Classification d'images satellites pour la cartographie de l'occupation/utilisation du sol

## Introduction

La classification est une tâche fondamentale en apprentissage automatique (Machine Learning). Elle consiste à attribuer une observation à une classe parmi un ensemble de catégories prédéfinies à partir de variables explicatives. En télédétection, elle est utilisée pour identifier des classes d’occupation du sol à partir d’images multispectrales, par exemple l’eau, la végétation, le sol nu ou les zones urbaines.

L’objectif est d’apprendre une règle de décision à partir d’exemples connus, puis de l’appliquer à de nouvelles observations. En pratique, on distingue principalement la classification supervisée, où les classes sont connues à l’avance, et les méthodes non supervisées, où l’algorithme découvre lui-même la structure des données.

## Classifications supervisée et non supervisée

### **Classification supervisée**

Dans l’apprentissage supervisé, chaque observation d’entraînement est associée à une classe cible. Le modèle apprend donc la relation entre les variables d’entrée et la classe à prédire. En classification d’images, cela signifie que les pixels ou objets d’apprentissage doivent être étiquetés au préalable.

Les algorithmes couramment utilisés incluent la régression logistique, les arbres de décision, Random Forest, les SVM et les réseaux de neurones. Ces méthodes sont particulièrement utiles lorsque les données étiquetées sont fiables et représentatives du problème étudié.

**Exemple de matrice d'apprentissage**

| Bleu | Vert | Rouge | NIR | SWIR1 | NDVI | NDBI | Classe |
|--------|--------|--------|--------|---------|---------|---------|---------|
| 0.05 | 0.04 | 0.03 | 0.02 | 0.01 | -0.20 | -0.35 | Eau |
| 0.08 | 0.12 | 0.09 | 0.60 | 0.18 | 0.74 | -0.28 | Végétation |
| 0.17 | 0.21 | 0.25 | 0.22 | 0.35 | -0.05 | 0.31 | Urbain |


### **Classification non supervisée**

Dans l’apprentissage non supervisé, aucune étiquette de classe n’est fournie. L’algorithme cherche alors à regrouper les observations similaires en clusters. Cette approche est utile pour explorer une structure inconnue dans les données, mais les groupes obtenus doivent ensuite être interprétés par l’utilisateur. Parmi les méthodes les plus connues, on trouve K-Means et le clustering hiérarchique. K-Means cherche à former des groupes compacts autour de centroïdes.

## Introduction au Machine Learning (Apprentissage auomatique)

Le **Machine Learning** regroupe un ensemble de méthodes qui permettent à un ordinateur d’apprendre à partir des données, sans être explicitement programmé règle par règle. L’objectif est de construire un modèle capable de détecter des relations, de faire des prédictions ou de révéler des structures utiles dans les données. On distingue généralement trois grands types d’apprentissage : supervisé, non supervisé et par renforcement. Chacun correspond à un type de problème différent et à une manière spécifique d’utiliser les données.

<img src="images/ml--2.svg" width="700">

### **Apprentissage supervisé**

Dans l’apprentissage supervisé, le modèle est entraîné à partir d’exemples dont la réponse est déjà connue. Chaque observation d’apprentissage possède donc une variable cible, ou label, que l’algorithme tente de prédire à partir des variables explicatives.

Les principales applications sont :

- Classification, lorsque la variable cible est qualitative.
- Régression, lorsque la variable cible est quantitative.

| Régression | Classification |
|------------|---------------|
| Variable continue | Variable catégorielle |
| Température | Type d'occupation du sol |
| Débit | Eau / Végétation / Urbain |
| Précipitations | Présence / Absence |

Exemples :

- Prévoir si un pixel d’image correspond à l’eau, à la végétation ou à une zone urbaine.
- Estimer un débit maximal, une température ou une concentration à partir de variables explicatives.

### **Apprentissage non supervisé**

Dans l’apprentissage non supervisé, aucune étiquette n’est fournie au modèle. L’algorithme cherche alors à découvrir des structures cachées, des groupes homogènes ou des relations internes dans les données. Les principales applications sont le **clustering**, pour regrouper les observations similaires, et la **réduction de dimension**, pour résumer l’information contenue dans un grand nombre de variables.

Exemples :

- Identifier des groupes de stations hydrologiques aux comportements proches.
- Résumer un grand jeu de variables spectrales par quelques composantes principales.

### **Apprentissage par renforcement**

Dans l’apprentissage par renforcement, un agent apprend à prendre des décisions en interagissant avec un environnement. À chaque action, il reçoit une récompense ou une pénalité, ce qui lui permet d’améliorer progressivement sa stratégie. Ce cadre est utilisé, par exemple, en robotique, en contrôle, en jeux, ou dans certains problèmes de décision séquentielle.

- Le jeu d’apprentissage, utilisé pour entraîner le modèle.
- Le jeu de test, utilisé uniquement pour évaluer ses performances sur des données jamais vues.

### **Variables explicatives et variable cible**

Dans tout problème d'apprentissage supervisé, il est nécessaire de distinguer deux types de variables :

- les **variables explicatives** (*features*) ;
- la **variable cible** (*target* ou *label*).

L'objectif du modèle est d'apprendre la relation existant entre les variables explicatives et la variable cible afin de pouvoir effectuer des prédictions sur de nouvelles données.

<img src="images/features-target.png" width="700">

#### **Variables explicatives (Features)**

Les variables explicatives correspondent aux informations utilisées par le modèle pour réaliser ses prédictions. En télédétection, ces variables sont généralement extraites des images satellites et peuvent inclure les bandes spectrales, les indices spectraux, des variables topographiques, des variables climatiques, des variables dérivées, etc.

#### **Variable cible (Target)**

La variable cible représente l'information que l'on souhaite prédire. Dans un problème de classification, il s'agit d'une variable catégorielle représentant une classe.

### **Biais et Variance**

L'un des principaux objectifs en Machine Learning est de construire un modèle capable de bien prédire de nouvelles observations. Pour comprendre pourquoi certains modèles fonctionnent mieux que d'autres, il est nécessaire d'introduire deux notions fondamentales : le **biais** et la **variance**. Ces deux concepts permettent d'expliquer les erreurs de prédiction d'un modèle et sont au cœur de la plupart des problèmes d'apprentissage automatique.

#### **Pourquoi parle-t-on de biais et de variance ?**

Lorsqu'un modèle est entraîné sur un jeu de données, il cherche à apprendre les relations existant entre les variables explicatives (*features*) et la variable cible (*target*). Cependant, un modèle peut commettre deux types d'erreurs :

- être trop simple et ne pas capturer correctement les relations présentes dans les données ;
- être trop complexe et apprendre des détails spécifiques au jeu d'entraînement qui ne se retrouveront pas dans de nouvelles données.

Ces deux situations sont respectivement associées au **biais** et à la **variance**.

---

#### **Le biais et le sous-ajustement (*underfitting*)**

Le biais mesure la capacité du modèle à représenter correctement les relations réelles présentes dans les données. Un modèle ayant un biais élevé est généralement trop simple et incapable de capturer la complexité du phénomène étudié. On dit alors que le modèle est **sous-ajusté** (*underfitting*).

<img src="images/underfitting.jpeg" width="500">


**Caractéristiques d'un modèle à fort biais :**

- modèle trop simple ;
- erreurs élevées sur les données d'entraînement ;
- erreurs élevées sur les données de test ;
- incapacité à apprendre les structures importantes des données.

**Exemples de modèles à fort biais :**

- régression linéaire pour un problème très non linéaire ;
- arbre de décision peu profond ;
- modèle utilisant trop peu de variables explicatives.

#### **La variance et le sur-ajustement (*overfitting*)**

La variance mesure la sensibilité du modèle aux données d'entraînement. Un modèle ayant une variance élevée s'adapte excessivement aux exemples observés et mémorise même les erreurs ou le bruit présents dans les données. On dit alors que le modèle est **sur-ajusté** (*overfitting*). Cela signifie que votre algorithme n'est pas en mesure de faire des prédictions précises : une modification même minime des données d'entrée entraîne un changement très important dans les résultats du modèle.

<img src="images/overfitting.png" width="400">

#### **Comparaison biais / variance**

| Biais élevé | Variance élevée |
|-------------|-----------------|
| Modèle trop simple | Modèle trop complexe |
| Sous-apprentissage | Sur-apprentissage |
| Mauvais sur train | Très bon sur train |
| Mauvais sur test | Mauvais sur test |
| Ne capture pas les relations | Mémorise le bruit |

#### **Le compromis biais-variance**

La qualité d’un modèle dépend souvent de l’équilibre entre biais et variance.

- Le biais correspond à l’erreur due à un modèle trop simplificateur.
- La variance correspond à la sensibilité du modèle aux variations du jeu d’apprentissage.

En pratique :

- Un biais élevé conduit souvent à de l’underfitting.
- Une variance élevée conduit souvent à de l’overfitting.

Le bon modèle est celui qui minimise l’erreur totale en trouvant un équilibre entre ces deux composantes.

### **Réglage des hyperparamètres (hyperparameter tunning)**

Lorsqu'on entraîne un modèle de Machine Learning, tous les paramètres ne sont pas obtenus de la même manière. Certains sont appris automatiquement à partir des données, tandis que d'autres doivent être définis avant même le début de l'apprentissage. Ces derniers sont appelés **hyperparamètres**. Les hyperparamètres jouent un rôle fondamental car ils déterminent la manière dont l'algorithme va apprendre. Ils influencent la structure du modèle, sa complexité, sa vitesse d'apprentissage et sa capacité à généraliser à de nouvelles données. En pratique, deux modèles entraînés sur exactement les mêmes données peuvent produire des résultats très différents simplement parce qu'ils utilisent des hyperparamètres différents.

#### **Paramètres et hyperparamètres : quelle différence ?**

Pour bien comprendre la notion d'hyperparamètre, il faut d'abord la distinguer de celle de paramètre.

Les **paramètres** sont les valeurs que le modèle apprend automatiquement pendant l'entraînement. Dans une régression linéaire, par exemple, les coefficients de la droite sont calculés à partir des données. Dans un arbre de décision, les seuils utilisés pour séparer les observations sont également déterminés automatiquement. Dans un réseau de neurones, les poids des connexions entre les neurones sont ajustés progressivement pendant l'apprentissage.

Les **hyperparamètres**, en revanche, ne sont pas appris à partir des données. Ils sont choisis avant l'entraînement et servent à guider le processus d'apprentissage. Ils définissent en quelque sorte les règles du jeu que le modèle devra suivre.

On peut faire une analogie avec un étudiant préparant un examen. Les connaissances acquises au cours de ses révisions correspondent aux paramètres du modèle. En revanche, le nombre d'heures consacrées à l'étude chaque jour, le choix des supports de cours ou la méthode de travail adoptée correspondent aux hyperparamètres. Ces choix influencent fortement les résultats obtenus, même s'ils ne constituent pas directement les connaissances apprises.

### **Évaluation d'un modèle de Machine Learning (*classificateur*)**

Construire un modèle de classification n'est qu'une première étape. Une fois le modèle entraîné, il est essentiel d'évaluer sa capacité à produire des prédictions fiables sur des données qu'il n'a jamais vues auparavant. L'objectif de l'évaluation n'est pas seulement de mesurer les performances d'un modèle, mais également de comprendre ses forces, ses faiblesses et sa capacité de généralisation. Un modèle qui obtient d'excellents résultats sur les données d'entraînement n'est pas nécessairement un bon modèle. Ce qui nous intéresse avant tout est sa performance sur de nouvelles observations.

#### **Évaluation sur un jeu de test**

Lors de l'entraînement, les données sont généralement divisées en deux ensembles :

- un ensemble d'entraînement (*training set*) ;
- un ensemble de test (*test set*).

Le modèle apprend uniquement à partir de l'ensemble d'entraînement. Une fois l'apprentissage terminé, il est évalué sur l'ensemble de test afin d'obtenir une estimation de ses performances sur de nouvelles données. Cette séparation permet d'éviter une erreur fréquente consistant à évaluer un modèle sur les mêmes données que celles utilisées pour son apprentissage. En télédétection, cela revient par exemple à entraîner un classificateur à partir d'un ensemble de pixels de référence puis à vérifier sa capacité à classer correctement des pixels indépendants.

#### **Validation croisée**

Les performances obtenues sur un unique découpage entraînement/test peuvent dépendre fortement des données sélectionnées. Pour obtenir une évaluation plus robuste, on utilise souvent la validation croisée (*cross-validation*). Le principe consiste à diviser les données en plusieurs sous-ensembles. À chaque itération, une partie des données est utilisée pour l'évaluation tandis que les autres servent à l'entraînement. Le processus est répété plusieurs fois de manière à ce que chaque observation soit utilisée pour le test une fois.

La méthode la plus courante est la validation croisée en *K plis* (*K-Fold Cross Validation*).

Par exemple, avec une validation croisée à 5 plis :

1. les données sont divisées en 5 groupes ;
2. 4 groupes servent à l'entraînement ;
3. le groupe restant sert à l'évaluation ;
4. l'opération est répétée 5 fois ;
5. les résultats sont ensuite moyennés.

Cette approche permet d'obtenir une estimation plus stable des performances du modèle.

#### Matrice de confusion

La matrice de confusion constitue l'un des outils les plus importants pour analyser les résultats d'un classificateur. Elle compare les classes observées aux classes prédites. Dans le cas d'une classification binaire, quatre situations sont possibles :

| Prédiction / Réalité | Réalité positive | Réalité négative |
|---|---|---|
| Prédit positif | Vrai Positif (VP) | Faux Positif (FP) |
| Prédit négatif | Faux Négatif (FN) | Vrai Négatif (VN) |

Cette matrice permet de comprendre les types d'erreurs commises par le modèle. En télédétection, elle permet par exemple d'identifier si des zones urbaines sont régulièrement confondues avec des sols nus ou si certaines surfaces en eau sont classées comme végétation.

#### Exactitude globale (Accuracy)

L'accuracy représente la proportion d'observations correctement classées parmi l'ensemble des observations.

Elle est définie par :

$$
Accuracy = \frac{VP + VN}{VP + VN + FP + FN}
$$

Cette métrique est simple à interpréter mais peut être trompeuse lorsque les classes sont déséquilibrées. Par exemple, si 95 % des pixels appartiennent à la classe "végétation", un modèle qui prédit systématiquement cette classe obtiendra une accuracy élevée tout en étant peu utile.

## Précision (Precision)

La précision mesure la fiabilité des prédictions positives réalisées par le modèle. Elle répond à la question :

> Parmi les observations classées dans une catégorie donnée, quelle proportion appartient réellement à cette catégorie ?

La précision est définie par :

$$
Precision = \frac{VP}{VP + FP}
$$

Une faible précision indique que le modèle produit de nombreux faux positifs.

## Rappel (Recall)

Le rappel mesure la capacité du modèle à détecter les observations appartenant réellement à une classe. Il répond à la question :

> Parmi toutes les observations réellement positives, combien ont été correctement identifiées ?

Le rappel est défini par :

$$
Recall = \frac{VP}{VP + FN}
$$

Un faible rappel signifie que de nombreuses observations sont manquées.

## F1-score

La précision et le rappel sont souvent complémentaires. Le F1-score combine ces deux métriques dans une mesure unique :

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

Cette métrique est particulièrement utile lorsque les classes sont déséquilibrées.

### **Algorithmes de classification courants dans Scikit-Learn**

Le choix d’un algorithme de classification dépend de la taille du jeu de données, du nombre de variables, du niveau de bruit, de la séparation entre les classes et de l’objectif d’interprétation. Certains modèles sont simples et faciles à expliquer, tandis que d’autres sont plus puissants mais plus sensibles au réglage des paramètres.

#### **K-Nearest Neighbors (KNN)**

Le **KNN** est une méthode simple qui classe une observation en fonction des $k$ observations les plus proches dans l’espace des variables. Il ne construit pas de modèle complexe à l’avance : la prédiction repose directement sur la distance entre les points.

<img src="images/knn.png" width="1500">

**Forces :**

- Simple à comprendre et à expliquer.
- Pas d’étape d’apprentissage complexe.
- Efficace sur des problèmes de petite taille.
- Peut être performant si les classes sont bien séparées.

**Faiblesses :**

- Sensible à l’échelle des variables, donc la standardisation est souvent nécessaire.
- Coûteux en prédiction sur de grands jeux de données.
- Sensible au bruit et aux observations aberrantes.
- Les performances peuvent se dégrader en grande dimension.

**Hyperparamètres principaux :**

- ***n_neighbors*** : nombre de voisins pris en compte.
- ***weights*** : pondération uniforme ou selon la distance.
- **metric** : métrique de distance utilisée.
- **algorithm** : méthode de recherche des voisins.

#### **Support Vector Machine (SVM)**

Le SVM cherche une frontière de décision qui sépare au mieux les classes en maximisant la marge entre elles. Il est particulièrement intéressant lorsque les classes sont bien structurées et que la dimension des variables est modérée à élevée.

<img src="images/svm.jpg" width="700">

**Forces :**

- Très performant sur de nombreux problèmes de classification.
- Efficace sur des jeux de données de taille modérée.
- Peut modéliser des séparations non linéaires grâce aux noyaux.
- Souvent robuste lorsqu’il est bien réglé.

**Faiblesses :**

- Nécessite souvent un bon prétraitement des variables.
- L’entraînement peut devenir coûteux sur de grands jeux de données.
- Moins interprétable que les modèles linéaires simples.
- Le choix du noyau et des paramètres influence fortement les résultats.

**Hyperparamètres principaux :**

- ***C*** : contrôle de la pénalisation des erreurs.
- ***kernel*** : type de noyau, par exemple linéaire, radial ou polynomial.
- ***gamma*** : influence d’un point d’entraînement pour les noyaux non linéaires.
- ***degree*** : degré du noyau polynomial.
- ***class_weight*** : utile en cas de classes déséquilibrées.

#### **Random Forest (RF)**

##### **Arbre de décision**

L’arbre de décision construit une suite de règles de séparation pour partitionner l’espace des variables et attribuer une classe à chaque observation. Il est intuitif, visuel et facile à interpréter.

<img src="images/Satellite-image-classification-decision-tree.png" width="700">

**Hyperparamètres principaux d'un arbre de décision :**

- ***max_depth*** : profondeur maximale de l’arbre.
- ***min_samples_split*** : nombre minimal d’observations pour couper un nœud.
- ***min_samples_leaf*** : nombre minimal d’observations dans une feuille.
- ***criterion*** : mesure utilisée pour choisir les séparations.
- ***max_features*** : nombre maximal de variables testées à chaque séparation.

##### **Le Random Forest : un modèle basé sur les arbres de décision**

L'algorithme **Random Forest** constitue une amélioration importante des arbres de décision (*Decision Trees*). Bien que les arbres de décision soient simples à comprendre et à interpréter, ils présentent une faiblesse majeure : ils ont tendance à **sur-apprendre** (*overfitting*) les données d'entraînement. Un arbre trop complexe peut mémoriser les exemples observés au lieu d'apprendre des règles générales, ce qui dégrade ses performances lorsqu'il est appliqué à de nouvelles données. Pour limiter ce problème, Random Forest repose sur l'idée de combiner un grand nombre d'arbres de décision plutôt que d'utiliser un seul arbre.

Le principe consiste à construire plusieurs arbres à partir de versions légèrement différentes du jeu de données. Chaque arbre est entraîné sur un sous-ensemble aléatoire des observations et sur un sous-ensemble aléatoire des variables explicatives. Cette stratégie permet d'obtenir des arbres différents les uns des autres et réduit le risque que tous commettent les mêmes erreurs.

Chaque arbre produit ensuite sa propre prédiction. Dans le cas d'un problème de classification, chaque arbre vote pour une classe. La classe obtenant le plus grand nombre de votes devient la prédiction finale du modèle.

<img src="images/rf.png" width="700">

Par exemple, si l'on souhaite classifier un pixel Landsat et que les prédictions de cinq arbres sont :

| Arbre | Classe prédite |
|---------|---------------|
| 1 | Végétation |
| 2 | Végétation |
| 3 | Eau |
| 4 | Végétation |
| 5 | Urbain |

La classe **Végétation** obtient la majorité des votes et sera donc retenue comme prédiction finale.

L'utilisation de nombreux arbres permet généralement d'obtenir des modèles plus robustes et plus précis qu'un arbre de décision unique. En effet, les erreurs commises par certains arbres sont souvent compensées par les prédictions correctes des autres. En télédétection, Random Forest est particulièrement populaire pour la classification de l'occupation du sol à partir d'images satellitaires. Il est capable de gérer simultanément un grand nombre de variables explicatives, telles que les bandes spectrales Landsat, les indices spectraux (NDVI, SAVI, MNDWI, LSWI, NDBI, etc.). Parmi ses principaux avantages figurent sa robustesse face au bruit, sa bonne capacité de généralisation et sa facilité d'utilisation. Il nécessite généralement peu de prétraitement des données et fournit également des informations sur l'importance des variables utilisées pour la classification. Pour ces raisons, Random Forest est aujourd'hui l'un des algorithmes les plus utilisés en télédétection, en cartographie de l'occupation du sol et dans de nombreuses applications environnementales.

**Forces :**

- Très bon pouvoir prédictif dans de nombreux contextes.
- Robuste au bruit et au surapprentissage.
- Gère bien des relations non linéaires.
- Fournit une estimation de l’importance des variables.

**Faiblesses :**

- Moins interprétable qu’un arbre unique.
- Peut être plus lourd en calcul qu’un arbre seul.
- Les importances de variables peuvent être biaisées dans certains cas.
- Les modèles très grands peuvent devenir coûteux à stocker et à prédire.

**Hyperparamètres principaux :**

- ***n_estimators*** : nombre d’arbres dans la forêt.
- ***max_depth*** : profondeur maximale des arbres.
- ***max_features*** : nombre de variables considérées à chaque séparation.
- ***min_samples_split*** : nombre minimal pour couper un nœud.
- ***min_samples_leaf*** : nombre minimal dans une feuille.
- ***bootstrap*** : échantillonnage avec remise ou non.
- ***class_weight*** : utile si les classes sont déséquilibrées.

##### **XGBoost**

Le Extreme Gradient Boosting (XGBoost) pour la classification est une méthode d’ensemble basée sur des arbres de décision construits de manière séquentielle. Chaque nouvel arbre apprend à corriger les erreurs des arbres précédents, ce qui améliore progressivement la qualité de la prédiction. Pour un problème de classification, XGBoost commence par une prédiction initiale simple, puis calcule les erreurs commises par le modèle. Ensuite, il entraîne un nouvel arbre sur ces erreurs, ajoute sa contribution au modèle précédent, et répète ce processus plusieurs fois jusqu’à obtenir un modèle final plus performant. XGBoost est performant parce qu’il combine plusieurs mécanismes : le boosting séquentiel, la minimisation d’une fonction de perte, et une régularisation qui limite la complexité du modèle.

<img src="images/xgboost.webp" width="700">

**Forces :**

- Excellentes performances prédictives dans de nombreux problèmes.
- Très efficace sur des données tabulaires.
- Capable de capturer des interactions complexes.
- Dispose d’outils de régularisation pour limiter l’overfitting.
- Peut fournir des mesures d’importance des variables.

**Faiblesses :**

- Plus complexe à régler qu’un modèle plus simple.
- Risque d’overfitting si les hyperparamètres sont mal choisis.
- Moins intuitif qu’un arbre de décision.
- Demande souvent un certain temps d’optimisation.

**Hyperparamètres principaux :**

- ***n_estimators*** : nombre d’arbres.
- ***learning_rate*** : vitesse d’apprentissage.
- ***max_depth*** : profondeur maximale des arbres.
- ***subsample*** : proportion des observations utilisées par arbre.
- ***colsample_bytree*** : proportion de variables utilisées par arbre.
- ***gamma*** : pénalisation pour effectuer une séparation.
- ***min_child_weight*** : poids minimal requis dans une feuille.
- ***reg_alpha*** : régularisation L1.
- ***reg_lambda*** : régularisation L2.

##### **Importance des variables**

Certains algorithmes permettent d’identifier les variables les plus influentes dans les prédictions. Cela est particulièrement utile en télédétection pour comprendre quelles bandes ou indices spectrals discriminent le mieux les classes. Par exemple, on peut évaluer si NDVI, MNDWI, NDBI, LSWI contribuent fortement à distinguer l’eau, la végétation, le sol nu et les zones urbaines. Les importances de variables sont particulièrement disponibles avec les arbres, Random Forest et XGBoost. Elles doivent cependant être interprétées avec prudence, car elles ne mesurent pas toujours l’effet causal réel des variables.

## Échantillonnage des pixels et préparation du jeu de données d'entraînement

Maintenant que les points d'entraînement ont été numérisés sous QGIS (avec un champ **`class`** indiquant la catégorie d'occupation du sol de chaque point) et que less **image composite Landsat** a été téléchargée, l'étape suivante consiste à **extraire la valeur des pixels** de l'image à l'emplacement de chaque point. C'est cette opération, appelée **échantillonnage** (*sampling*), qui permet de construire une **matrice d'apprentissage exploitable par un algorithme de Machine Learning** : chaque ligne correspond à un point d'entraînement, chaque colonne correspond à une bande spectrale (ou un indice), et une colonne supplémentaire contient la classe à prédire.

```
Point d'entraînement (x, y, classe)  +  Image multibande
                    ↓
   Valeur des bandes au pixel correspondant
                    ↓
   Bleu | Vert | Rouge | NIR | SWIR1 | SWIR2 | Classe
```

Les étapes ci-dessous permettent de :

1. Charger l'image composite (`.tif`) et les points d'entraînement (`.gpkg`).
2. Vérifier que les deux jeux de données partagent le même système de coordonnées (CRS).
3. Échantillonner les valeurs des bandes à l'emplacement de chaque point.
4. Construire une matrice d'apprentissage (`DataFrame`) propre, sans valeurs manquantes.
5. Séparer les données en un jeu d'entraînement et un jeu de test.
6. Entraîner un premier modèle de classification (Random Forest) et évaluer ses performances.

### Classification de l'image de 2025

#### Librairies

En plus des bibliothèques déjà utilisées (`os`, `geopandas`), nous avons besoin de :

- `rasterio` pour lire l'image raster et échantillonner les valeurs des pixels,
- `numpy` et `pandas` pour manipuler les données sous forme de tableaux,
- `scikit-learn` pour la séparation entraînement/test, l'entraînement du modèle et son évaluation.

````python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import rasterio
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

#### Charger l'image et les points d'entraînement

- `image_path` : l'image composite multibande produite précédemment.
- `training_path` : la couche de points d'entraînement créée sous QGIS, contenant le champ `class`.

Le CRS des points doit obligatoirement correspondre à celui de l'image. Si ce n'est pas le cas, l'échantillonnage renverrait des valeurs incorrectes (les points seraient projetés au mauvais endroit sur l'image). On utilise donc `to_crs()` pour aligner systématiquement les points sur le CRS du raster.

````python
training_path = 'data/classification_training_dataset/training_dataset.gpkg'
image_path = 'data/images_landsat/L08_2025.tif'

# Charger les points d'entraînement
training_points = gpd.read_file(training_path)
training_points = training_points.dropna(subset=['class'])
training_points['class'] = training_points['class'].astype(int)

# Aligner le CRS des points sur celui de l'image
with rasterio.open(image_path) as src:
    raster_crs = src.crs

training_points = training_points.to_crs(raster_crs)

print(f"Nombre de points d'entraînement : {len(training_points)}")
training_points.head()
````

#### Échantillonnage des valeurs des pixels

La méthode `rasterio.sample()` permet d'extraire, pour une liste de coordonnées `(x, y)`, la valeur de chaque bande du raster au pixel correspondant. On construit ensuite un `DataFrame` où chaque ligne correspond à un point, chaque colonne à une bande, et on y ajoute la colonne `classe` issue des points d'entraînement.

> ⚠️ L'ordre des points dans `coords` doit correspondre exactement à l'ordre des lignes de `training_points`, afin que la colonne `classe` corresponde bien aux bonnes valeurs de pixels.

````python
# Liste des coordonnées (x, y) de chaque point
coords = [(geom.x, geom.y) for geom in training_points.geometry]

with rasterio.open(image_path) as src:
    band_count = src.count
    band_names = [src.descriptions[i] if src.descriptions[i] else f"B{i+1}" for i in range(band_count)]
    sampled_values = list(src.sample(coords))
````

La colonne `training_points['class']` contient actuellement les codes numériques 1, 2, 3, 4, on peut les remplacer par leurs noms avant d'effectuer l'échantillonage.

````python
# Dictionnaire de correspondance
class_mapping = {
    1: "Eau",
    2: "Végétation",
    3: "Bâti",
    4: "Sol nu"
}

# Construction de la matrice d'apprentissage
samples_df = pd.DataFrame(sampled_values, columns=band_names)
samples_df['class'] = training_points['class'].map(class_mapping)

samples_df.head()
````

#### Nettoyage des données

Certains points peuvent tomber sur des pixels sans données valides (par exemple un pixel masqué par les nuages dans le composite, ou un point situé légèrement en dehors de l'emprise de l'image). Ces lignes contiennent des valeurs manquantes (`NaN`) ou des valeurs aberrantes et doivent être retirées avant l'entraînement, sous peine de faire planter le modèle ou de fausser l'apprentissage.

````python
# Remplacer les valeurs infinies éventuelles, puis retirer les lignes incomplètes
samples_df2 = samples_df.replace([np.inf, -np.inf], np.nan).dropna()

print(f"Nombre de points valides après nettoyage : {len(samples_df)}")
samples_df2['class'].value_counts()
````

#### Séparation des variables explicatives et de la variable cible

Conformément à la distinction présentée dans la section théorique, on sépare :

- `X` : les **variables explicatives** (les bandes spectrales),
- `y` : la **variable cible** (la classe d'occupation du sol).

On divise ensuite les données en un **jeu d'entraînement** (utilisé pour ajuster le modèle) et un **jeu de test** (utilisé uniquement pour évaluer ses performances sur des données non vues). Le paramètre `stratify=y` permet de conserver les mêmes proportions de classes dans les deux jeux, ce qui est important lorsque certaines classes sont moins représentées que d'autres.

````python
X = samples_df2.drop(columns='class')
y = samples_df2['class']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print(f"Taille du jeu d'entraînement : {len(X_train)}")
print(f"Taille du jeu de test : {len(X_test)}")
````

### Entraînement d'un premier modèle : Random Forest

On entraîne un classificateur **Random Forest**, présenté dans la section théorique pour sa robustesse et sa popularité en classification d'images satellites. `n_estimators` fixe le nombre d'arbres de la forêt ; les autres hyperparamètres pourront être ajustés ultérieurement (par exemple via une validation croisée).

````python
rf_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
````

### Évaluation du modèle sur le jeu de test

On applique le modèle entraîné au jeu de test (jamais vu pendant l'apprentissage) afin d'obtenir une estimation honnête de ses performances. On calcule l'**exactitude globale** (*accuracy*), ainsi que la **précision**, le **rappel** et le **F1-score** pour chaque classe via `classification_report`, et on affiche la **matrice de confusion** afin d'identifier les classes les plus souvent confondues entre elles (par exemple sol nu / urbain).

````python
y_pred = rf_model.predict(X_test)

print("Exactitude globale (accuracy) :", accuracy_score(y_test, y_pred))
print()
print("Rapport de classification :")
print(classification_report(y_test, y_pred))
print()
print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))
````

#### Interprétation des performances du modèle Random Forest

Le modèle Random Forest atteint une **exactitude globale (accuracy) de 84,3 %**, ce qui signifie que 59 observations sur 70 ont été correctement classées. Cette amélioration nette par rapport aux versions précédentes indique une meilleure capacité de séparation des classes d’occupation du sol.

##### Analyse par classe

Les performances varient selon les classes :

- **Végétation** est également très bien classée avec un F1-score de **0,89**. Le modèle la reconnaît de manière fiable (recall = 85 %).
- **Sol nu** présente de bonnes performances avec un F1-score de **0,80**, mais quelques confusions apparaissent avec la classe Bâti (6 cas).
- **Bâti** reste la classe la plus difficile à classifier(F1-score = **0,74**, recall = 0,87)

### Optimisation des hyperparamètres du Random Forest

Le modèle entraîné à l'étape précédente utilise les hyperparamètres par défaut de scikit-learn (à l'exception de `n_estimators=300`). Ces valeurs ne sont pas nécessairement optimales pour notre jeu de données. L'optimisation des hyperparamètres consiste à explorer systématiquement un espace de configurations possibles et à retenir celle qui maximise les performances mesurées par **validation croisée** sur le jeu d'entraînement.

Nous allons utiliser deux stratégies complémentaires :

1. **`RandomizedSearchCV`** : explore un sous-ensemble aléatoire de l'espace des hyperparamètres. Très efficace quand cet espace est large.
2. **`GridSearchCV`** : explore exhaustivement toutes les combinaisons d'un sous-ensemble restreint, centré autour des meilleurs résultats trouvés par la recherche aléatoire.

Cette approche en deux étapes — souvent appelée ***coarse-to-fine*** — permet de couvrir un large espace de recherche tout en restant raisonnable en temps de calcul.

```
Espace large  →  RandomizedSearchCV  →  Zone prometteuse  →  GridSearchCV  →  Meilleurs hyperparamètres
```


#### Librairies supplémentaires

On ajoute `RandomizedSearchCV` et `GridSearchCV` pour la recherche d'hyperparamètres, ainsi que `matplotlib` et `seaborn` pour les visualisations.

````python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, f1_score
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

#### Modèle de référence (hyperparamètres par défaut)

Avant toute optimisation, on enregistre les performances du modèle déjà entraîné avec les hyperparamètres par défaut. Ce score servira de **référence** (*baseline*) pour mesurer le gain apporté par l'optimisation.

````python
# Modèle baseline (déjà entraîné)

y_pred_baseline = rf_model.predict(X_test)
acc_baseline = accuracy_score(y_test, y_pred_baseline)

print(f"Accuracy baseline (hyperparamètres par défaut) : {acc_baseline:.4f}")
print()
print(classification_report(y_test, y_pred_baseline))
````

#### Recherche aléatoire (RandomizedSearchCV)

On définit un **espace de recherche large** couvrant les hyperparamètres les plus influents du Random Forest :

| Hyperparamètre | Rôle |
|---|---|
| `n_estimators` | Nombre d'arbres dans la forêt |
| `max_depth` | Profondeur maximale de chaque arbre |
| `min_samples_split` | Nombre minimal d'observations pour diviser un nœud |
| `min_samples_leaf` | Nombre minimal d'observations dans une feuille terminale |
| `max_features` | Nombre de variables testées à chaque séparation |
| `bootstrap` | Échantillonnage avec ou sans remise |

`RandomizedSearchCV` tire aléatoirement `n_iter` combinaisons dans cet espace et évalue chacune par validation croisée à `cv` plis.

````python
from scipy.stats import randint

param_dist = {
    'n_estimators'    : randint(100, 600),
    'max_depth'       : [None, 5, 10, 20, 30, 40],
    'min_samples_split': randint(2, 20),
    'min_samples_leaf' : randint(1, 10),
    'max_features'    : ['sqrt', 'log2', 0.3, 0.5, None],
    'bootstrap'       : [True, False],
}

random_search = RandomizedSearchCV(
    estimator  = RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions = param_dist,
    n_iter     = 50,
    cv         = 5,
    scoring    = 'accuracy',
    random_state = 42,
    n_jobs     = -1,
    verbose    = 1,
    return_train_score = True,
)

random_search.fit(X_train, y_train)

print("\nMeilleure combinaison trouvée :")
print(random_search.best_params_)
print(f"\nMeilleure accuracy (CV) : {random_search.best_score_:.4f}")
````

#### Évaluation du modèle optimisé

On applique le modèle aux meilleurs hyperparamètres sur le jeu de test et on enregistre ses performances.

````python
best_rf = random_search.best_estimator_
y_pred_tuned = best_rf.predict(X_test)
acc_tuned = accuracy_score(y_test, y_pred_tuned)

print(f"Accuracy modèle optimisé : {acc_tuned:.4f}")
print()
print(classification_report(y_test, y_pred_tuned))
````

#### Tableau de synthèse — Hyperparamètres et performances

Un récapitulatif des configurations et performances des deux modèles : le **tableau de synthèse** permet de comparer directement les hyperparamètres retenus par la recherche avec les valeurs par défaut, et de quantifier le gain d'accuracy.

````python
best_params = random_search.best_params_

summary = pd.DataFrame({
    'Hyperparamètre': list(best_params.keys()),
    'Baseline (défaut)': [
        rf_model.get_params()[k] for k in best_params.keys()
    ],
    'Optimisé': list(best_params.values()),
})

best_rf      = random_search.best_estimator_
y_pred_tuned = best_rf.predict(X_test)
acc_tuned    = accuracy_score(y_test, y_pred_tuned)

print("═" * 60)
print("  SYNTHÈSE — OPTIMISATION DES HYPERPARAMÈTRES")
print("═" * 60)
print()
print(summary.to_string(index=False))
print()
print("─" * 60)
print(f"  Accuracy baseline  : {acc_baseline:.4f}")
print(f"  Accuracy optimisée : {acc_tuned:.4f}")
print(f"  Gain               : {(acc_tuned - acc_baseline)*100:+.4f}% ({(acc_tuned - acc_baseline)*100:+.2f} pp)")
print("─" * 60)
````

> Le modèle optimisé `best_rf` est prêt à être utilisé pour classifier l'ensemble de l'image.


### Classification de l'ensemble de l'image

Une fois le modèle optimisé, on l'applique à **l'intégralité de l'image** pour produire une carte de classification pixel par pixel. La difficulté principale est d'ordre mémoire : une image Landsat peut contenir des dizaines de millions de pixels, qu'il serait dangereux de charger tous en mémoire d'un coup.

On adopte donc un traitement **par blocs** (*tiled prediction*) :

```
Image multibande  →  Bloc de N lignes  →  Reshape en (N×cols, bands)
                  →  Prédiction RF      →  Reshape en (N, cols)
                  →  Écriture dans le raster de sortie
```

Cette approche permet de classifier des images de n'importe quelle taille quelle que soit la mémoire disponible.


````python
import rasterio
from rasterio.windows import Window
import numpy as np

output_path = 'data/classification/classification_2015.tif'
os.makedirs(os.path.dirname(output_path), exist_ok=True)

class_map = {
    "Eau": 1,
    "Végétation": 2,
    "Bâti": 3,
    "Sol nu": 4
}

NODATA_VAL = 255

with rasterio.open(image_path) as src:
    profile = src.profile.copy()
    n_bands = src.count
    band_names = [src.descriptions[i] if src.descriptions[i] else f"B{i+1}" 
                  for i in range(n_bands)]

    # lecture image complète
    img = src.read().astype(np.float32)   # (bands, rows, cols)

# reshape vers pixels
rows, cols = img.shape[1], img.shape[2]
pixels = img.reshape(n_bands, -1).T

# masque validité
valid = ~np.any(np.isnan(pixels) | (pixels == 0), axis=1)

# résultat
result = np.full(pixels.shape[0], NODATA_VAL, dtype=np.uint8)

if valid.sum() > 0:
    pred = best_rf.predict(pd.DataFrame(pixels[valid], columns=band_names))
    pred = np.array([class_map[p] for p in pred])
    result[valid] = pred.astype(np.uint8)

# reshape image
result_img = result.reshape(rows, cols)

# sauvegarde GeoTIFF
profile.update(count=1, dtype="uint8", nodata=NODATA_VAL, compress="lzw")

with rasterio.open(output_path, "w", **profile) as dst:
    dst.write(result_img, 1)

print(f"✓ Classification terminée → {output_path}")
````

#### Vérification rapide de la sortie

On relit le raster de classification pour vérifier sa structure et afficher la distribution des classes prédites (nombre de pixels par classe).

````python
with rasterio.open(output_path) as src:
    classif = src.read(1)
    pixel_size_m = abs(src.transform.a)   # résolution spatiale en mètres
    crs_classif  = src.crs
    nodata_val   = src.nodata

# Distribution des classes (on exclut nodata)
valid_pixels = classif[classif != NODATA_VAL]
unique, counts = np.unique(valid_pixels, return_counts=True)

class_names = {
    1: "Eau",
    2: "Végétation",
    3: "Bâti",
    4: "Sol nu"
}

print(f"Résolution spatiale : {pixel_size_m} m")
print(f"CRS : {crs_classif}")
print(f"Dimensions : {classif.shape}  ({classif.shape[0]*classif.shape[1]:,} pixels au total)")
print(f"Pixels valides classifiés : {len(valid_pixels):,}")
print()
print("Distribution des classes :")
for cls, cnt in zip(unique, counts):
    print(f"  {class_names.get(cls, 'Inconnu')} ({cls}) : {cnt:>10,} pixels")
````

````python
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap, BoundaryNorm
import matplotlib.pyplot as plt
import numpy as np
import rasterio

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 14

CLASS_INFO = {
    1: {'label': 'Eau',        'color': "#75b5dc"},
    2: {'label': 'Végétation', 'color': "#2b5b40"},
    3: {'label': 'Bâti',       'color': "#d06b66"},
    4: {'label': 'Sol nu',     'color': "#eacfb1"},
    255: {'label': 'NoData',   'color': 'black'}
}

class_ids = sorted([k for k in CLASS_INFO.keys() if k != 255])
colors = [CLASS_INFO[c]['color'] for c in class_ids]

cmap_classif = ListedColormap(colors)
norm_classif = BoundaryNorm(
    boundaries=[0.5, 1.5, 2.5, 3.5, 4.5],
    ncolors=len(class_ids)
)

with rasterio.open(output_path) as src:
    classif = src.read(1)
    extent_img = [src.bounds.left, src.bounds.right,
                  src.bounds.bottom, src.bounds.top]

# masque nodata
classif_plot = classif.astype(float)
classif_plot[classif == 255] = np.nan

fig, ax = plt.subplots(figsize=(12, 10))

im = ax.imshow(
    classif_plot,
    extent=extent_img,
    origin='upper',
    cmap=cmap_classif,
    norm=norm_classif,
    interpolation='none'
)

legend_handles = [
    Patch(facecolor=CLASS_INFO[c]['color'],
          edgecolor='white',
          label=CLASS_INFO[c]['label'])
    for c in class_ids
]

ax.legend(handles=legend_handles,
          title='Classes',
          loc='lower right',
          framealpha=0.9)

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlabel("")
ax.set_ylabel("")

plt.tight_layout()

out_path = "resultats/classification/carte_classification_2015.png"
os.makedirs(os.path.dirname(out_path), exist_ok=True)
plt.savefig(out_path, dpi=300)

print("✓ Carte sauvegardée → carte_classification.png")
````

#### Calcul des superficies par classe

La superficie occupée par chaque classe se déduit directement du **nombre de pixels** classifiés multiplié par la **surface élémentaire d'un pixel** (résolution spatiale²).

$$
\text{Superficie}_c = N_c \times (\text{résolution})^2
$$

On exprime les résultats en **hectares** (1 ha = 10 000 m²) et en **km²**, et on en déduit les **proportions** de chaque classe sur la zone d'étude.


````python
pixel_area_m2 = pixel_size_m ** 2          # m²  par pixel
pixel_area_ha = pixel_area_m2 / 10_000     # ha  par pixel
pixel_area_km2= pixel_area_m2 / 1_000_000  # km² par pixel

rows_area = []
for cls_id in class_ids:
    n_pix = int(np.sum(classif == cls_id))
    rows_area.append({
        'Classe'          : cls_id,
        'Label'           : CLASS_INFO[cls_id]['label'],
        'Pixels'          : n_pix,
        'Superficie (ha)' : round(n_pix * pixel_area_ha, 2),
        'Superficie (km²)': round(n_pix * pixel_area_km2, 4),
    })

area_df = pd.DataFrame(rows_area)
total_ha  = area_df['Superficie (ha)'].sum()
total_km2 = area_df['Superficie (km²)'].sum()
area_df['Proportion (%)'] = (area_df['Superficie (ha)'] / total_ha * 100).round(2)

area_df_display = area_df.copy()
area_df_display['Pixels']           = area_df_display['Pixels'].apply(lambda x: f'{x:,}')
area_df_display['Superficie (ha)']  = area_df_display['Superficie (ha)'].apply(lambda x: f'{x:,.2f}')
area_df_display['Superficie (km²)'] = area_df_display['Superficie (km²)'].apply(lambda x: f'{x:,.4f}')
area_df_display['Proportion (%)']   = area_df_display['Proportion (%)'].apply(lambda x: f'{x:.2f} %')

print(f"Résolution spatiale : {pixel_size_m} m  →  {pixel_area_ha:.4f} ha/pixel")
print(f"Superficie totale classifiée : {total_ha:,.2f} ha  ({total_km2:,.2f} km²)")
print()
display(area_df_display.set_index('Classe'))
````

#### Vectorisation de la carte de classification (optionnel)

La carte raster peut être **vectorisée** (convertie en polygones) avec `rasterio.features.shapes()`. Cela permet d'exporter la classification en **GeoPackage** (.gpkg) ou Shapefile pour l'ouvrir directement dans QGIS, d'appliquer des requêtes attributaires, ou d'intégrer les résultats dans un SIG existant.

````python
from rasterio.features import shapes
from shapely.geometry import shape

vector_output = 'resultats/classification/classification_2015.gpkg'

with rasterio.open(output_path) as src:
    classif_arr = src.read(1)
    transform   = src.transform
    crs_out     = src.crs

# Vectorisation pixel → polygones
mask = classif_arr != NODATA_VAL
geom_list = []
for geom_dict, class_val in shapes(classif_arr, mask=mask, transform=transform):
    geom_list.append({
        'geometry' : shape(geom_dict),
        'class'    : int(class_val),
        'label'    : CLASS_INFO.get(int(class_val), {}).get('label', str(int(class_val))),
    })

gdf_classif = gpd.GeoDataFrame(geom_list, crs=crs_out)

# Superficie en ha (en reprojetant en métrique si nécessaire)
gdf_classif_proj = gdf_classif.to_crs(gdf_classif.estimate_utm_crs())
gdf_classif['superficie_ha'] = (gdf_classif_proj.geometry.area / 10_000).round(4)

gdf_dissolved = gdf_classif.dissolve(by="class", as_index=False)

gdf_dissolved.to_file(vector_output, driver='GPKG', layer='classification_2015')

print(f"✓ Couche vectorielle exportée → {vector_output}")
print(f"  {len(gdf_dissolved):,} polygones")
gdf_dissolved
````
