**Python**, créé par **Guido van Rossum en 1991**, est un langage de programmation polyvalent, facile à apprendre, simple et très lisible. Contrairement à une idée très répandue, **Python** ne tire pas son nom du serpent. Son créateur, **Guido van Rossum**, était fan de l'émission humoristique britannique [Monty Python's Flying Circus](https://fr.wikipedia.org/wiki/Monty_Python%27s_Flying_Circus). Lorsqu'il a développé le langage à la fin des années 1980, il cherchait un nom court, facile à retenir et un peu décalé. Il a donc choisi *Python* en référence aux **Monty Python**, et non à l'animal.

### Les variables

Une **variable** est comme une boîte étiquetée dans laquelle on range une information, pour pouvoir la réutiliser plus tard sans avoir à la réécrire. En Python, on crée une variable simplement en lui donnant un nom, suivi du signe `=`, suivi de la valeur que l'on souhaite y stocker :

Par exemple :

````python
prenom = "Léa"
age = 20

print(prenom)
print(age)
````

On peut aussi combiner plusieurs variables dans un même `print()` :

````python
print("Je m'appelle", prenom, "et j'ai", age, "ans.")
````

Une fois créée, une variable peut être **réutilisée**, **modifiée**, ou utilisée dans des calculs :

````python
age = age + 1  # on ajoute 1 à l'âge
print("L'année prochaine, j'aurai", age, "ans.")
````

> **📌 À retenir**
>
> - Le nom d'une variable doit commencer par une lettre ou un underscore (`_`), ne pas contenir d'espace, et ne pas être un mot réservé de Python (comme `print`).
> - On privilégie des noms de variables **explicites** : `age` plutôt que `a`, `prenom` plutôt que `x`.
> - En Python, les noms de variables s'écrivent traditionnellement en minuscules, avec des underscores pour séparer les mots (`mon_prenom`, `taille_max`).


### Les types de variables

Chaque variable contient une information d'un certain **type**. Les quatre types de base à connaître pour débuter sont :

| Type | Nom Python | Exemple | Description |
|---|---|---|---|
| Entier | `int` | `20` | Un nombre sans virgule |
| Décimal | `float` | `1.75` | Un nombre à virgule |
| Chaîne de caractères | `str` | `"Léa"` | Du texte, entre guillemets |
| Booléen | `bool` | `True` / `False` | Une valeur Vrai/Faux |

Pour connaître le type d'une variable, on utilise la fonction `type()` :

````python
prenom = "Léa"
age = 20
taille = 1.75
est_etudiante = True

print(type(prenom))
print(type(age))
print(type(taille))
print(type(est_etudiante))
````

> **⚠️ Attention**
>
> Le type d'une variable dépend de la façon dont on écrit sa valeur :
>
> - `20` (sans guillemets) → c'est un **entier** (`int`) ;
> - `"20"` (avec guillemets) → c'est une **chaîne de caractères** (`str`), même si ça ressemble à un nombre !
>
> Cette nuance est une source d'erreur très fréquente chez les débutants. Le code ci-dessous illustre le problème :

````python
a = 5        # un entier (int)
b = "5"      # une chaîne de caractères (str)

print(a + a)   # 5 + 5 -> addition, donne 10
print(b + b)   # "5" + "5" -> concaténation de texte, donne "55"
````

On peut **convertir** une variable d'un type vers un autre grâce aux fonctions `int()`, `float()`, `str()` :

````python
age_texte = "20"           # ceci est une chaîne de caractères
age_nombre = int(age_texte)  # on la convertit en entier

print(age_nombre + 1)
print(type(age_nombre))
````

### Les listes

Une **liste** permet de regrouper plusieurs valeurs dans une seule variable, par exemple les notes d'un étudiant ou les noms des villes d'un département. En Python, une liste s'écrit entre crochets `[ ]`, avec les éléments séparés par des virgules.

````python
villes = ["Dakar", "Thies", "Saint-Louis", "Mbour"]
print(villes)
print(type(villes))
````

Chaque élément d'une liste possède une **position**, appelée *index*. En Python, **on compte à partir de 0** (et non de 1) :


````python
print(villes[0])  # premier élément
print(villes[1])  # deuxième élément
print(villes[-1]) # dernier élément
````

On peut modifier un élément, en ajouter un nouveau, ou compter le nombre d'éléments :


````python
villes[1] = "Kaolack"      # on remplace "Lyon" par "Bordeaux"
villes.append("Ziguinchor")  # on ajoute un élément à la fin

print(villes)
print("Nombre de villes :", len(villes))
````

> **📌 À retenir**
>
> - Une liste est **ordonnée** (l'ordre des éléments est conservé) et **modifiable** (on peut la changer après l'avoir créée).
> - On accède à un élément avec `liste[index]`, en commençant à `0`.
> - Quelques méthodes utiles : `.append(x)` (ajouter à la fin), `.remove(x)` (supprimer un élément), `len(liste)` (compter les éléments).


### Les tuples

Un **tuple** ressemble beaucoup à une liste : c'est aussi une collection ordonnée de valeurs. La différence essentielle est qu'un tuple s'écrit entre **parenthèses** `( )` et qu'il est **immuable** : une fois créé, on ne peut plus le modifier.


````python
coordonnees_dakar = (14.8566, -14.3522)  # (latitude, longitude)

print(coordonnees_dakar)
print("Latitude :", coordonnees_dakar[0])
print("Longitude :", coordonnees_dakar[1])
````

````python
# Essayer de modifier un tuple provoque une erreur :
coordonnees_dakar[0] = 50.0
````

> **📌 À retenir**
>
> - On utilise un **tuple** quand on veut être sûr que les valeurs ne changeront jamais (des coordonnées géographiques, une date, etc.).
> - On utilise une **liste** quand la collection est amenée à évoluer (ajout, suppression, modification).

### Les dictionnaires

Un **dictionnaire** stocke des informations sous forme de paires **clé : valeur**, un peu comme un vrai dictionnaire associe un mot à sa définition. Il s'écrit entre accolades `{ }`.

````python
etudiant = {
    "prenom": "Astou",
    "age": 20,
    "filiere": "Géographie"
}

print(etudiant)
print(type(etudiant))
````

On accède à une valeur grâce à sa **clé** (et non à un index numérique comme pour les listes) :

````python
print(etudiant["prenom"])
print(etudiant["filiere"])
````

On peut ajouter une nouvelle paire clé/valeur, ou modifier une valeur existante :

````python
etudiant["ville"] = "Sedhiou"   # ajout d'une nouvelle clé
etudiant["age"] = 21          # modification d'une valeur existante

print(etudiant)
````

> **📌 À retenir**
>
> - Un dictionnaire associe des **clés uniques** à des **valeurs** : `{clé: valeur}`.
> - Très utile pour représenter une fiche d'information structurée (un étudiant, une ville, un produit...).
> - `.keys()` liste les clés, `.values()` liste les valeurs, `.items()` liste les paires clé/valeur.

### Les conditions

Les **conditions** permettent à un programme de prendre des décisions : *"si telle situation se produit, alors fais ceci ; sinon, fais autre chose"*. En Python, on utilise les mots-clés `if`, `elif` (contraction de *else if*) et `else`.

On compare des valeurs grâce aux opérateurs suivants :

| Opérateur | Signification |
|---|---|
| `==` | égal à |
| `!=` | différent de |
| `>` / `<` | supérieur à / inférieur à |
| `>=` / `<=` | supérieur ou égal à / inférieur ou égal à |

````python
age = 20

if age >= 18:
    print("Vous êtes majeur.")
else:
    print("Vous êtes mineur.")
````

````python
note = 14

if note >= 16:
    print("Mention Très Bien")
elif note >= 14:
    print("Mention Bien")
elif note >= 10:
    print("Admis(e)")
else:
    print("Non admis(e)")
````

> **⚠️ Attention**
>
> En Python, l'indentation (les espaces en début de ligne) n'est pas une question d'esthétique : elle fait partie du langage ! Le bloc de code à exécuter sous un `if`, un `elif` ou un `else` **doit** être indenté (généralement avec 4 espaces), sinon Python renvoie une erreur.

### Les fonctions

Une **fonction** est un bloc de code réutilisable, auquel on donne un nom, et que l'on peut exécuter (on dit "appeler") autant de fois que nécessaire, avec des valeurs différentes à chaque fois. Vous en utilisez déjà depuis le début de ce tutoriel : `print()`, `len()`, `type()` et `int()` sont toutes des fonctions !

On définit une fonction avec le mot-clé `def`, suivi de son nom, de ses **paramètres** entre parenthèses, et du mot-clé `return` pour indiquer ce qu'elle doit renvoyer :

````python
def saluer(prenom):
    message = "Bonjour " + prenom + ", bienvenue en cours de Python !"
    return message


print(saluer("Ibrahima"))
print(saluer("Aicha"))
````

Une fonction peut prendre plusieurs paramètres, et faire des calculs :

````python
def calculer_imc(poids, taille):
    imc = poids / (taille ** 2)
    return imc


mon_imc = calculer_imc(68, 1.75)
print("IMC :", round(mon_imc, 1))
````

> **📌 À retenir**
>
> - Une fonction se **définit** une fois (`def`) et peut être **appelée** autant de fois que nécessaire.
> - Les éléments entre parenthèses sont les **paramètres** (les informations dont la fonction a besoin pour fonctionner).
> - `return` indique ce que la fonction renvoie comme résultat ; sans `return`, la fonction ne renvoie rien.
> - Les fonctions évitent de répéter plusieurs fois le même code : c'est un des piliers d'une bonne programmation.


### Les modules

Un **module** est un fichier contenant des fonctions déjà écrites par d'autres, que l'on peut importer et réutiliser dans son propre code, plutôt que de tout réécrire soi-même. Python est fourni avec de nombreux modules intégrés (la *bibliothèque standard*), et on peut en installer beaucoup d'autres (NumPy, Pandas, Geopandas...), comme vous l'avez fait dans l'environnement `td_l3_geo`.

On importe un module avec le mot-clé `import` :


````python
import math

print(math.sqrt(16))   # racine carrée
print(math.pi)         # la constante π
````

On peut aussi importer une seule fonction d'un module avec `from ... import ...`, ou donner un **alias** (un surnom plus court) à un module avec `as` :

````python
from random import randint

nombre_aleatoire = randint(1, 100)  # un nombre entier au hasard entre 1 et 100
print(nombre_aleatoire)
````

### Les DataFrames (avec Pandas)

Quand on travaille avec des données réelles (un tableau de relevés météo, une liste de communes avec leur population...), on utilise rarement des listes ou des dictionnaires "à la main". On utilise plutôt **Pandas**, une bibliothèque externe spécialisée dans la manipulation de données tabulaires (comme un tableau Excel).

L'objet central de Pandas est le **DataFrame** : un tableau composé de lignes et de colonnes, où chaque colonne peut contenir un type de donnée différent.

On peut créer un DataFrame à partir d'un dictionnaire, où chaque clé devient une colonne :

````python
import pandas as pd

data = {
    "ville": ["Paris", "Lyon", "Marseille", "Lille"],
    "population": [2148000, 522000, 870000, 233000],
    "superficie_km2": [105.4, 47.9, 240.6, 34.7]
}

df = pd.DataFrame(data)
df
````

````python
df.columns       # liste les noms des colonnes
````

### Les tableaux avec NumPy (arrays)

**NumPy** est la bibliothèque de base du calcul scientifique en Python. Elle introduit un nouvel objet, le **array** (tableau), qui ressemble à une liste mais est beaucoup plus rapide et adapté aux calculs sur de grandes quantités de nombres. C'est la fondation sur laquelle reposent Pandas, Geopandas, Rasterio ou encore Xarray.

On importe NumPy avec l'alias conventionnel `np` :

````python
import numpy as np

temperatures = np.array([12.5, 14.0, 15.2, 13.8, 16.1])

print(temperatures)
print(type(temperatures))
````

La grande force des arrays NumPy est de permettre des **calculs vectorisés** : une opération s'applique automatiquement à tous les éléments, sans avoir besoin d'écrire de boucle.

````python
# Convertir toutes les températures de Celsius en Fahrenheit, d'un seul coup
temperatures_f = temperatures * 9/5 + 32
print(temperatures_f)

print("Moyenne :", temperatures.mean())
print("Minimum :", temperatures.min())
print("Maximum :", temperatures.max())
````

Un array peut avoir plusieurs dimensions : un tableau à une dimension (comme ci-dessus), à deux dimensions (une grille, comme une image en noir et blanc), voire plus. On parle alors de sa **forme** (`shape`) :

````python
grille = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(grille)
print("Forme (lignes, colonnes) :", grille.shape)
````

> **📌 À retenir**
>
> - Un **array NumPy** est une grille de valeurs (1D, 2D, 3D...), toutes du **même type**.
> - Les opérations (`+`, `-`, `*`, `/`) s'appliquent **élément par élément**, sans boucle à écrire.
> - `.shape` donne les dimensions du tableau, ce qui sera essentiel pour manipuler des images satellites (lignes × colonnes × bandes).

### Introduction à Xarray

**Xarray** prolonge l'idée des arrays NumPy en y ajoutant des **étiquettes** : au lieu de manipuler des données uniquement par leur position (ligne 3, colonne 5...), on peut les manipuler par leurs **coordonnées réelles** (une date, une latitude, une longitude...). C'est très pratique pour les données géospatiales et climatiques, qui ont souvent plusieurs dimensions (temps, x, y, bandes...).

L'objet de base de Xarray est le **`DataArray`** :

````python
import xarray as xr
import numpy as np

# Températures pour 3 jours, sur une petite grille de 2 x 2 points
valeurs = np.array([
    [[12.5, 13.0], 
     [11.8, 12.2]],

    [[14.1, 14.6], 
     [13.5, 13.9]],
     
    [[15.0, 15.4], 
     [14.2, 14.8]],
])

temperature = xr.DataArray(
    valeurs,
    dims=["temps", "y", "x"],
    coords={
        "temps": ["2026-06-01", "2026-06-02", "2026-06-03"],
        "y": [45.0, 45.1],
        "x": [3.0, 3.1],
    },
    name="temperature"
)

temperature
````

Grâce à ces étiquettes, on peut sélectionner les données par leur **nom de dimension**, plutôt que par leur position, ce qui rend le code beaucoup plus lisible :

````python
# Toutes les valeurs du premier jour
temperature.sel(temps="2026-06-01")
````

````python
# Moyenne dans le temps, pour chaque point de la grille
temperature.mean(dim="temps")
````

> **📌 À retenir**
>
> - Un **`DataArray`** Xarray est un array NumPy enrichi de **dimensions nommées** (`dims`) et de **coordonnées** (`coords`).
> - Un **`Dataset`** Xarray regroupe plusieurs `DataArray` qui partagent les mêmes coordonnées (par exemple : température **et** précipitations, sur la même grille).
> - C'est l'outil que vous utiliserez pour manipuler des séries d'images satellites (une dimension *temps* en plus des dimensions spatiales *x* et *y*), notamment via **`rioxarray`**.

### Lire une image avec Rasterio

Une image satellite (ou tout autre fichier raster, comme un `.tif`) est en réalité une grille de valeurs numériques, accompagnée d'informations de **géoréférencement** (où se situe cette grille sur Terre, avec quelle résolution, dans quel système de coordonnées...). **Rasterio** est la bibliothèque de référence en Python pour lire, écrire et manipuler ce type de fichiers.

Pour découvrir Rasterio sans avoir besoin d'une vraie image satellite, commençons par **créer** un petit fichier raster fictif, que nous lirons ensuite :

````python
import numpy as np
import rasterio
from rasterio.transform import from_origin

# Une petite grille 10 x 10, avec un dégradé de valeurs (par exemple une élévation fictive)
donnees = np.arange(100, dtype="float32").reshape(10, 10)

# Géoréférencement fictif : coin supérieur gauche, résolution de 30 m
transform = from_origin(west=650000, north=6600000, xsize=30, ysize=30)

with rasterio.open(
    "exemple_raster.tif", "w",
    driver="GTiff",
    height=donnees.shape[0],
    width=donnees.shape[1],
    count=1,
    dtype=donnees.dtype,
    crs="EPSG:2154",  # Lambert 93
    transform=transform,
) as fichier:
    fichier.write(donnees, 1)

print("Fichier 'exemple_raster.tif' créé avec succès.")
````

On peut maintenant **ouvrir** ce fichier avec Rasterio, exactement comme on ouvrirait une vraie image satellite téléchargée :

````python
with rasterio.open("exemple_raster.tif") as image:
    print("Dimensions (largeur x hauteur) :", image.width, "x", image.height)
    print("Nombre de bandes :", image.count)
    print("Système de coordonnées :", image.crs)
    print("Emprise géographique :", image.bounds)

    bande_1 = image.read(1)  # on lit la première bande sous forme d'array NumPy

print("\nType de la bande lue :", type(bande_1))
print("Forme :", bande_1.shape)
````

````python
import matplotlib.pyplot as plt

plt.imshow(bande_1, cmap="terrain")
plt.colorbar(label="Valeur")
plt.title("Aperçu de notre raster fictif (élévation (m))")
plt.show()
````

> **📌 À retenir**
>
> - Un raster ouvert avec `rasterio.open()` donne accès aux **métadonnées** (dimensions, nombre de bandes, CRS, emprise) et aux **valeurs** via `.read()`.
> - `.read(1)` renvoie un **array NumPy** : tout ce que vous avez appris sur NumPy s'applique directement aux images satellites !
> - On utilise `with rasterio.open(...) as image:` pour que le fichier soit automatiquement et proprement fermé après utilisation.

### Lire une image avec Rioxarray

**Rioxarray** combine le meilleur des deux mondes : il s'appuie sur Rasterio pour lire les fichiers raster, mais renvoie le résultat sous forme d'un objet **Xarray**, avec dimensions nommées, coordonnées géographiques et géoréférencement intégré. C'est l'outil que vous utiliserez le plus souvent dans ce cours pour manipuler des images satellites.

Reprenons notre fichier `exemple_raster.tif` créé plus haut :

````python
import rioxarray

image_xr = rioxarray.open_rasterio("exemple_raster.tif")
image_xr
````

On retrouve la structure d'un objet Xarray vue précédemment, avec en plus les informations de géoréférencement directement accessibles :

````python
print("Dimensions :", image_xr.dims)
print("Système de coordonnées :", image_xr.rio.crs)
print("Résolution :", image_xr.rio.resolution())
````

L'affichage est également très simple grâce à la méthode `.plot()`, intégrée à Xarray :

````python
image_xr.plot(cmap="terrain")
````

> **📌 À retenir**
>
> - `rioxarray.open_rasterio()` ouvre un fichier raster directement sous forme d'objet **Xarray géoréférencé**.
> - On accède aux informations spatiales via l'**accesseur** `.rio` : `.rio.crs`, `.rio.resolution()`, `.rio.reproject()`, `.rio.clip()`...
> - En résumé : **Rasterio** pour une lecture/écriture bas niveau et performante, **Rioxarray** pour une manipulation haut niveau, étiquetée, et qui s'intègre parfaitement avec Xarray et Pandas/Geopandas.
